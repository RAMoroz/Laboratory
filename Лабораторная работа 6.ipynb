Функция леммитизации

```python
from string import punctuation
def lemmatize(input_text):
    tokens = nltk.word_tokenize(input_text)
    normed_tokens=[morph.parse(s)[0].normal_form for s in tokens]
    #исключаем стоп-слова - предлоги, союзы, междометья
    normed_tokens=[word for word in normed_tokens if word not in nltk.corpus.stopwords.words('Russian')]
    #исключаем знаки пунктуации
    normed_tokens=[word for word in normed_tokens if word not in punctuation]
    return ' '.join(normed_tokens)
```

Формирование датафрейма с  текстами

```python
import os
import pandas as pd

#пустой датафрейм
df=pd.DataFrame(columns=['Text','Class'])
#Директории текстов
dir0="Lab6/Acoustics"
dir1="Lab6/History"

#считаем все наши тексты в датафрейм с указанием класса
for filename in os.listdir(dir0):
    with open(os.path.join(dir0,filename),encoding="utf8") as file:
        contents = lemmatize(file.read())
    df = df.append(pd.Series({'Text':contents, 'Class':0}), ignore_index=True)
```

```python
for filename in os.listdir(dir1):
    with open(os.path.join(dir1,filename),encoding="utf8") as file:
        contents = lemmatize(file.read())
    df = df.append(pd.Series({'Text':contents, 'Class':1}), ignore_index=True)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train, y_test=train_test_split(df['Text'],df['Class'],test_size=0.3,random_state=64)#250
```

Bag-of-Words

```python
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
bof_vect=CountVectorizer()
bof_vect.fit(np.hstack([X_train,X_test]))
bof_train=bof_vect.transform(X_train)
bof_test=bof_vect.transform(X_test)
```

```python
bof_train.toarray()
```

```python
bof_train.toarray().shape
```

TF - IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vect=TfidfVectorizer()
tfidf_vect.fit(np.hstack([X_train,X_test]))
tfidf_train=tfidf_vect.transform(X_train)
tfidf_test=tfidf_vect.transform(X_test)
```

```python
tfidf_train.toarray()
tfidf_train.toarray().shape
```

Word2Vec

```python
from gensim.models import Word2Vec
X_train_w2v=X_train.apply(str.split)
X_test_w2v=X_test.apply(str.split)
w2v_vect=Word2Vec(np.hstack([X_train_w2v,X_test_w2v]),size=300,min_count=1,workers=8) #min_count=1
```

```python
X_train_w2v
```

```python
#X_train_w2v[12]
```

```python
def text2vec(text):
    vecs=[]
    for word in text:
        try:
            vecs.append(w2v_vect[word])
        except KeyError:
            pass
        return np.sum(vecs, axis=0)/len(vecs)
w2v_train = X_train_w2v.apply(text2vec)
w2v_test = X_test_w2v.apply(text2vec)
w2v_train
```

```python
w2v_train.shape
```

```python
w2v_train = np.dstack(w2v_train)[0]
w2v_train.shape
```

```python
w2v_test = np.dstack(w2v_test)[0]
```

Классификация

BOF

```python
Acoustics_mean_bof=np.sum(bof_train[y_train==0],axis=0)
Acoustics_mean_bof.shape
```

```python
History_mean_bof=np.sum(bof_train[y_train == 1],axis=0)
```

TF - IDF

```python
Acoustics_mean_tfidf=np.sum(tfidf_train[y_train==0],axis=0)
Acoustics_mean_tfidf.shape
```

```python
History_mean_tfidf=np.sum(tfidf_train[y_train == 1],axis=0)
```

Word2Vec

```python
Acoustics_mean_w2v=np.sum(w2v_train[:,y_train == 0],axis=1)
Acoustics_mean_w2v.shape
```

```python
History_mean_w2v=np.sum(w2v_train[:,y_train == 1],axis=1)
```

Посмотрим на их вид

```python
Acoustics_mean_bof
```

```python
Acoustics_mean_tfidf
```

```python
Acoustics_mean_w2v
```

```python
from scipy.spatial.distance import cosine
bof_Acoustics = np.apply_along_axis(cosine,1,bof_test.toarray(),v=Acoustics_mean_bof)
bof_History = np.apply_along_axis(cosine,1,bof_test.toarray(),v=History_mean_bof)
bof_results = pd.DataFrame([
    bof_Acoustics,
    bof_History,
    np.maximum(bof_Acoustics,bof_History) == bof_Acoustics,
    y_test
],index=["Acoustics","History","Predict","Class"]).T.astype(np.float)
bof_results
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(bof_results['Predict'],bof_results['Class'])
```

```python
tfidf_Acoustics = np.apply_along_axis(cosine,1,bof_test.toarray(),v=Acoustics_mean_tfidf)
tfidf_History = np.apply_along_axis(cosine,1,bof_test.toarray(),v=History_mean_tfidf)
tfidf_results = pd.DataFrame([
    tfidf_Acoustics,
    tfidf_History,
    np.maximum(tfidf_Acoustics,tfidf_History) == tfidf_Acoustics,
    y_test
],index=["Acoustics","History","Predict","Class"]).T.astype(np.float)
tfidf_results
```

```python
accuracy_score(tfidf_results['Predict'],tfidf_results['Class'])
```

```python
w2v_Acoustics = np.apply_along_axis(cosine,0,w2v_test,v=Acoustics_mean_w2v)
w2v_History = np.apply_along_axis(cosine,0,w2v_test,v=History_mean_w2v)
w2v_results = pd.DataFrame([
    w2v_Acoustics,
    w2v_History,
    np.maximum(w2v_Acoustics,w2v_History) == w2v_Acoustics,
    y_test
],index=["Acoustics","History","Predict","Class"]).T.astype(np.float)
w2v_results
```

```python
accuracy_score(w2v_results['Predict'],w2v_results['Class'])
```
